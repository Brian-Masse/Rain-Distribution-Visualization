```python
import pandas as pd
import altair as alt

alt.data_transformers.disable_max_rows()
```

```python
xls = pd.ExcelFile(
    "./data/WeatherData.xlsx"
)
data = pd.read_excel( xls, "WeatherData" )
chart = alt.Chart(data)

months = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]

investigating_data = "tempRounded"
```

```python
def axis_config(sizes):
    return alt.Axis(
        domainColor="#7D82D4"+sizes[2],
        grid=False,
        ticks=False,
        labelColor="#ffffff",
        labelPadding=5,
        labelFontSize=sizes[0],
        titleColor="#ffffff",
        titlePadding=5,
        titleFontSize=sizes[1],
    )
```

```python
dates = pd.to_datetime( data["datetime"] )

def slice_data(month):
    index = months.index( month ) + 1
    rows = dates.dt.month == index
    month = data.loc[ rows, [investigating_data] ]
    return month

def create_Chart(month):
    month_data = slice_data( month )

    x_axis_sizes = (0, 0, "")
    y_axis_sizes = (0, 0, "00")
    if month == "dec":
        x_axis_sizes = (10, 15, "")
        y_axis_sizes = (10, 15, "00")

    chart = alt.Chart( month_data ).mark_area(
        filled=True,
        fill="#91B1F680",
        strokeWidth=5

    ).encode(
        alt.X( investigating_data + ":Q",
        title="Temperature (F)",
        scale=alt.Scale( 
            domainMax=100,
            domainMin=-5
        ),
        axis= axis_config(x_axis_sizes)
        ),
        alt.Y( "count()",
        title="Count of Temp in an hour",
        scale=alt.Scale( 
            domainMax=200,
            domainMin=0
        ), axis= axis_config(y_axis_sizes)
        ),
        ).properties(
            title=month,
            height=50
        )

    line = chart.mark_line(
        fill="#7D82D4"
    )
    
    return chart + line
```

```python
main = create_Chart( "jan" ) & create_Chart( "feb" ) & create_Chart( "mar" ) & create_Chart( "apr" ) & create_Chart( "may" ) & create_Chart( "jun" ) & create_Chart( "jul" ) & create_Chart( "aug" ) & create_Chart( "sep" ) & create_Chart( "oct" ) & create_Chart( "nov" ) & create_Chart( "dec" )

alt.renderers.set_embed_options(
    padding={"left": 0, "right": 0, "bottom": 0, "top": 0}
)

main.properties(
    title="The Temperature Distributions In Andover from 2021 - 2022 "
).configure(
    font="DINCondensed-Bold",
    background="#2C264A"
).configure_view(
    strokeWidth=0,
    # fill="#A896EB80"
).configure_title(
    fontSize=20,
    color="#FFFFFF"
)
```
